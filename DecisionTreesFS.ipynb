```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/shanksghub/FootballStats/master/FootballStats.csv'
df = pd.read_csv(url)
# Dataset is now stored in a Pandas Dataframe
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.drop(['appearances','hit woordwork','crosses','assists','name','blocks','fouls','goals conceded','headed clearances','clearances','saves','punches','yellow cards','red cards','offsides','clean sheets'],axis=1)

```

```python
df.describe()
```

```python
df = df.dropna(axis = 0, how ='any') 
```

```python
>>> df.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('position',axis=1)
y = df['position']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
>>> df.fillna(0)
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from IPython.display import Image  
from sklearn.externals.six import StringIO  
from sklearn.tree import export_graphviz
import pydot 

features = list(df.columns[1:])
features
```

```python
dot_data = StringIO()  
export_graphviz(dtree, out_file=dot_data,feature_names=features,filled=True,rounded=True)

graph = pydot.graph_from_dot_data(dot_data.getvalue())  
Image(graph[0].create_png())  
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfc_pred))
```

```python
print(classification_report(y_test,rfc_pred))
```
