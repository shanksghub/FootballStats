```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


url = 'https://raw.githubusercontent.com/shanksghub/FootballStats/master/FootballStats.csv'
df = pd.read_csv(url)

sns.lmplot(x='goals', y='assists', data=df)
```

```python
# Scatterplot arguments
sns.lmplot(x='goals', y='passes', data=df,fit_reg=True)   # Color by evolution stage


```
