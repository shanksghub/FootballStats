```
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor as dtr
from sklearn import tree
from sklearn.model_selection import GridSearchCV, train_test_split
```

```
url = 'https://raw.githubusercontent.com/shanksghub/FootballStats/master/FootballStats.csv'

df1 = pd.read_csv(url)
```

```
df1.describe()
df1 = df1.drop(["position","appearances","blocks","fouls","goals conceded","headed clearances","clearances","saves","punches","yellow cards","red cards","offsides","clean sheets"],axis=1)
```

```
df1["goals"].fillna(df1["goals"].mean(), inplace=True)
```

```
df1
```

```
df1["goals"] = df1["goals"].apply(np.log) 
y = df1["goals"]
df1 = df1.drop("goals",axis=1)
df1
```

```
x_train,x_test,y_train,y_test = train_test_split(df1,y,test_size = 0.2)
```

```
regressor = dtr(random_state=0)
```

```
y_pred = regressor.fit(x_train,y_train)
y_pred
```
